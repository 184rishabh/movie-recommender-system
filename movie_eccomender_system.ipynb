```python
import pandas as pd
```

```python
movies=pd.read_csv('movies.csv')
```

```python
ratings=pd.read_csv('ratings.csv')
```

```python
movies
```

```python
ratings
```

```python
df=pd.merge(movies,ratings)
```

```python
df
```

```python
df=df.drop(['genres','timestamp'],axis='columns')
```

```python
df
```

```python
user_rating=df.pivot_table(index=['userId'],columns=['title'],values='rating')
```

```python
user_rating
```

```python
df=user_rating
```

```python
df.fillna(0)
```

```python
#movie with less than ten rating are removed
```

```python
df=df.dropna(thresh=10,axis='columns')
```

```python
df
```

```python
df=df.fillna(0)
```

```python
similarity=df.corr(method='pearson')
```

```python
similarity.head(710)
```

```python
def recommendation(movie_name):
    x=similarity[movie_name]
    x=x.sort_values(ascending=False)
    x=pd.DataFrame(x)
    return x.head()
```

```python
r=recommendation("Fast & Furious (Fast and the Furious 4, The) (2009)")
```
